```python
students_details
```

### TASK

* Create a new column and name it Eligibility and it need to return 1 for Age>17 and 0 if Age<17.

```python
def get_age(x):
    if x>17:
        return 1
    else:
        return 0
```

```python
students_details['Eligibility'] = students_details['age'].apply(get_age)
```

```python
students_details
```

```python
students_details['Elibility_2'] =  students_details['age'].apply(lambda x: 1 if x>17 else 0)
```

```python
students_details
```

## Chapter 8 - Grouping Vs Pivottable Vs Crosstab

```python
insurance_data = pd.read_csv('insurance.csv')
insurance_data
```

```python
insurance_data.shape
```

```python
insurance_data.isna().sum()
```

```python
insurance_data.describe(include = 'all')
```

### 1. Find out the region-wise average insurance charges.

```python
insurance_data['region'].unique()
```

```python
insurance_data.groupby( by=['region'])['charges'].mean().sort_values(ascending = False)
```

### 1. Find out the smoker with regionwise average insurance charges.

```python
insurance_data.groupby( by=['region','smoker'])['charges'].mean().sort_values(ascending = False)
```

```python
pd.pivot_table(data = insurance_data,values='charges',index='region',columns='smoker',aggfunc='mean')
```

```python
pd.crosstab(index = insurance_data['smoker'],columns = insurance_data['region'],margins=True) #frequency table of the factors - it should be categorical factor
```
